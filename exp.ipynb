```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
import pickle
```

```python
dset = pd.read_csv('Churn_Modelling.csv')
```

```python
dset
```

```python
df = dset.drop(['RowNumber', 'CustomerId', 'Surname'], axis='columns')
```

```python
df
```

```python
le = LabelEncoder()
```

```python
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
geo = ohe.fit_transform(df[['Geography']])
```

```python
ohe.get_feature_names_out()
```

```python
df_encod = pd.DataFrame(geo, columns=ohe.get_feature_names_out())
```

```python
df_encod
```

```python
df1 = pd.concat([df.drop('Geography', axis='columns'), df_encod], axis='columns')
```

```python
df1
```

```python
with open('le.pkl', 'wb') as f:
    pickle.dump(le,f)
```

```python
with open('ohe.pkl', 'wb') as f:
    pickle.dump(ohe,f)
```

```python
x = df1.drop('Exited', axis='columns')
```

```python
y = df1['Exited']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
import datetime
```

```python
x_train.shape
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
ls = tf.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=opt, loss=ls, metrics=['accuracy'])
```

```python
log_dir = 'logs/fit' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
tf_cback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
hist = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs = 100,
    callbacks = [tf_cback, es]
)
```

```python
model.save('model.h5')
```
